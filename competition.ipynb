```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
le = LabelEncoder()
ord = OrdinalEncoder()
df_train['location_type'] = le.fit_transform(df_train['location_type'])
df_train['land_use'] = le.fit_transform(df_train['land_use'])
df_train['zoning_code'] = ord.fit_transform(df_train[['zoning_code']])
df_test['location_type'] = le.fit_transform(df_test['location_type'])
df_test['land_use'] = le.fit_transform(df_test['land_use'])
df_test['zoning_code'] = ord.fit_transform(df_test[['zoning_code']])
```

```python
df_train.drop('id',axis=1,inplace=True)
df_test.drop('id',axis=1,inplace=True)
```

```python
df_train_hand_outliers=df_train[numerical_cols]
df_train_hand_outliers.shape
```

```python
def handle_outliers(df, columns):
    df = df.copy()
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR


        df[col] = np.where(df[col] > upper_limit, Q3, df[col])
        df[col] = np.where(df[col] < lower_limit, Q1, df[col])
    return df
```

```python
df_train_hand_outliers = handle_outliers(df_train,df_train_hand_outliers)
```
